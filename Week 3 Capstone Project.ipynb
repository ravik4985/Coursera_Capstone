# Using Geocoder to get geographical coridinate 

```python
!pip install geopy
from geopy.geocoders import Nominatim # module to convert an address into latitude and longitude values
```

```python
loc=pd.read_csv('http://cocl.us/Geospatial_data')
```

```python
nghood=pd.merge(df,loc, on='Postal Code')
nghood
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronot are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(nghood['Latitude'], nghood['Longitude'], nghood['Borough'], nghood['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

```python
k=5
toronto_clustering = nf.drop(['Postal Code', 'Borough', 'Neighbourhood',],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
labels=kmeans.labels_
```

```python
nf['Labels']=labels
nf
```
